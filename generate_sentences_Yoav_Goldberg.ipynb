```python
import pandas as pd
import numpy as np
# Read dataframe
df = pd.read_csv('malayalam_aligned.tsv', sep='\t', encoding='utf-8')
df.shape, df.columns, df['sentID'].unique().shape
```

```python
sentToremove = df[df['vib'].isnull()]['sentID'].unique()
df = df[~(df['sentID'].isin(sentToremove))]
df.shape, df['sentID'].unique().shape
```

```python
# Remove sentences with null vib value
cl = ['Unnamed: 0', 'sentID', 'word', 'lemma', 'DEPS', 'vib']
lemma_dep_group_by = df[cl].groupby(['lemma', 'DEPS'])
lemma_group_by = df[cl].groupby('lemma')
```

```python
df.index
```

```python
# For a lemma and dep value, find its possible nonce vib values
# First get a set of vibs co-occurring with a (lemma, dep) : lemma_dep_to_vib = {}
# Then get a set of all vibs co-occurring with a lemma : lemma_to_vib = {}

lemma_dep_to_vib = {}
lemma_to_vib = {}

def get_possible_replacements(lemma, dep, lemma_dep_group_by, lemma_group_by):
    
    lp_set = set(list(lemma_dep_group_by.get_group((lemma, dep))['vib']))
    
    l_set = set(list(lemma_group_by.get_group((lemma))['vib']))
    
    possible_vibs = l_set - lp_set
    possible_vibs = possible_vibs - set(['SAME'])
    return possible_vibs

for name, group in lemma_dep_group_by:
    lemma_dep_to_vib[name] = get_possible_replacements(name[0], name[1], lemma_dep_group_by, lemma_group_by)
# print(lemma_dep_to_vib)
```

```python
deps_to_include = ['k1', 'k2', 'k3', 'k4', 'k5', 'r6', 'k7', 'vmod', 'nmod']
count = 0
for key, val in lemma_dep_to_vib.items():
    if len(val) > 40 and key[1] in set(deps_to_include):
        print(key)
        print(val)
        count +=1
print(count)
lemma_vib_group_by = df[cl].groupby(['lemma', 'vib'])
sentID_group_obj = df.groupby('sentID')
sent_ids = list(sentID_group_obj.groups.keys())
```

```python
# Iterate over each of the sentences and get nonce sentences, only for the given dependency types


sentences_dict = {}

traversed = {}

max_nonce = 10

incorrect_traversed = {}

nonce_dict = {}

def get_nonce(sid):
    
    if sid not in traversed:
        traversed[sid] = set()
    
    
    # Indices to words for each sentence (type: Int64Index([]))
    sentence_index = sentID_group_obj.groups[sid]
    sentence_df = df[df.index.isin(sentence_index)]
    
    index_list = []
    i = 0
    for index, row in sentence_df.iterrows():
        
        index_list.append(index)
        
        if row['DEPS'] not in set(deps_to_include):
            i +=1
            continue
        if row['vib'] == 'SAME':
            i +=1
            continue
        else:
            # Get possible replacements
            if row['DEPS'] not in nonce_dict:
                nonce_dict[row['DEPS']] = set()
            replacements = lemma_dep_to_vib[(row['lemma'], row['DEPS'])]
            if len(replacements) == 0:
                continue
            eligible_df= df[(df['lemma'] == row['lemma']) & (df['vib'].isin(list(replacements)))]
            nonce_dict[str(row['DEPS'])].add((sid, index, row['Unnamed: 0'], i, str(list(eligible_df['Unnamed: 0']))))
    
        i +=1
for n, sid in enumerate(sent_ids):
    if n %1000 == 0:
        print("Sentences Done : {0}/{1}".format(n, len(sent_ids)))
    if sid not in sentences_dict:
        sentences_dict[sid] = []
    get_nonce(sid)

nonce_count_dict = {}
count = 0
for key, val in nonce_dict.items():
    nonce_count_dict[key] = len(val)
    count += len(val)
nonce_count_dict
```

```python
max_example_from_one_sent = 10

nonce_deps_sents_dict = {}
for key, val in nonce_dict.items():
    for v in list(val):
        if len(v) == 0:
            continue
        sid = v[0]
        index = v[1]
        unnamed_index = v[2]
        index_i = v[3]
        total_index_replacements = v[4][1:-1].split(',')
        if len(total_index_replacements) == 0:
            continue
        for n in range(len(total_index_replacements)):
            index_to_replace = int(v[4][1:-1].split(',')[n])
            sentence_index = sentID_group_obj.groups[sid]
            text = list(df[df.index.isin(sentence_index)]['word'])
            nonce_text = text.copy()
            nonce_text[index_i] = list(df[df['Unnamed: 0'] == index_to_replace]['word'])[0]
            if key not in nonce_deps_sents_dict:
                nonce_deps_sents_dict[key] = []
            nonce_deps_sents_dict[key].append((text, nonce_text, sid, index, unnamed_index, index_i, index_to_replace))
            
            if n >= max_example_from_one_sent:
                break
```

```python
nonce_deps_sents_dict_count = {}
count = 0
for key, val in nonce_deps_sents_dict.items():
    nonce_deps_sents_dict_count[key] = len(val)
    count += len(val)
nonce_deps_sents_dict_count
# nonce_deps_sents_dict['k3']
```
